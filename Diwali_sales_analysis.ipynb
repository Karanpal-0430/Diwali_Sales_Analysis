```python
df.columns
```

```python
ax = sns.countplot(x ='Gender' ,data=df)
sns.set(rc={"figure.figsize":(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen =df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(x= "Gender" ,y= "Amount" ,data=sales_gen)
sns.set(rc={"figure.figsize":(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

form the above grapth we can see that most of the buyers are females and even the purchasing power of the females are greater then men

```python
display(HTML('<span style="font-size: 20px; font-weight: bold;">Age</span>'))
```

```python
ax=sns.countplot(data= df,x="Age Group", hue="Gender")
sns.set(rc={"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Most AMOUNT IN AGE GROUP
Sales_Age=df.groupby(["Age Group"], as_index=False).sum().sort_values(by="Amount" ,ascending=False)
sns.barplot(x="Age Group",y="Amount" ,data=Sales_Age,palette="Spectral",hue="Age Group")
sns.set(rc={"figure.figsize":(10,5)})
```

From the above graph we can see most of the buyers are aged 26-35 Female


```python
#Total Oreder/Sales of top 10 States
Sales_State=df.groupby(["State"], as_index=False).sum().sort_values(by="Orders" ,ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="State", y="Orders" ,data=Sales_State,palette="rainbow",hue="State")
```

```python
#Total Amount/Sales of top 10 States
Sales_State=df.groupby(["State"], as_index=False).sum().sort_values(by="Amount" ,ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="State", y="Amount" ,data=Sales_State,palette="rainbow",hue="State")
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

```python
ax=sns.countplot(data=df, x="Marital_Status")
sns.set(rc={"figure.figsize":(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_State=df.groupby(["Marital_Status","Gender"], as_index=False).sum().sort_values(by="Amount" ,ascending=False)
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(x="Marital_Status", y="Amount" ,data=Sales_State,palette="rainbow",hue="Gender")
```

From the above Graph we can see that most of the buyer are married (women) they have the most purchasing power

```python
display(HTML('<span style="font-size: 20px; font-weight: bold;">Occupation</span>'))
```

```python
ax=sns.countplot(data= df,x="Occupation", hue="Occupation")
sns.set(rc={"figure.figsize":(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_State=df.groupby(["Occupation"], as_index=False).sum().sort_values(by="Amount" ,ascending=False)
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(x="Occupation", y="Amount" ,data=Sales_State,palette="rainbow",hue="Occupation")
```

From above graphs we can see that most of the buyers are working in IT,Aviation and Healthcare sector

```python
ax=sns.countplot(data= df,x="Product_Category", hue="Product_Category")
sns.set(rc={"figure.figsize":(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_State=df.groupby(["Product_Category"], as_index=False).sum().sort_values(by="Amount" ,ascending=False).head(10)
sns.set(rc={"figure.figsize":(30,10)})
sns.barplot(x="Product_Category", y="Amount" ,data=Sales_State,palette="rainbow",hue="Product_Category")
```

from the above graph we can see the most sold product is food,clothing and electronics

```python
Sales_State=df.groupby(["Product_Category"], as_index=False).sum().sort_values(by="Orders" ,ascending=False).head(10)
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(x="Product_Category", y="Orders" ,data=Sales_State,palette="rainbow",hue="Product_Category")
```

from the above graph we can see the most sold product is clothing,food and electronics
